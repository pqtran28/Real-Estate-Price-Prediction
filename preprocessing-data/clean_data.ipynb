```python
def chuan_phaply(x: str):
    x = x.capitalize()
    if x == 'Hợp đồng mua bán':
        x = 'Hợp đồng'
    elif x not in ['Sổ đỏ', 'Sổ hồng', 'Hợp đồng', 'Sổ đỏ/sổ hồng']:
        x = 'Không rõ'
    return x
```

```python
df['phap_ly_chuan'] = df['phap_ly_chuan'].apply(lambda x: chuan_phaply(x) if pd.notna(x) else x)
df['phap_ly_chuan'].unique()
```
